```python
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5])  
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

subplot function is used to create and arrange multiple subplot within a single matplotlib figure.

```python
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title("line 1")
```

```python
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title("line 2")
```

```python
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title("line 3")
```

```python
plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title("line 4")
```

quation 4:

A bar plot, also known as a bar chart or bar graph, is a graphical representation of data using rectangular bars of varying lengths or heights. Each bar typically represents a category or group, and the length or height of the bar is proportional to the value it represents.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
```

```python
plt.barh(company,profit)
```

quation 5:
box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It is used to summarize the central tendency, spread, and potential outliers in the data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot(box1)
plt.title("box1")
```

```python
plt.boxplot(box2)
plt.title("box 2")
```
